# Listen to Pipeline Outputs

Play back **main_female**, **main_male**, **child**, and **background** WAVs produced by **Stage 03 ├втВмтАЭ Speaker Classification**.

A random participant is chosen automatically. Re-run the setup cell to pick a different one, or set `PARTICIPANT_ID` manually.

```python
import random
from pathlib import Path
import IPython.display as ipd
import soundfile as sf

# ├втАЭтВм├втАЭтВм Configuration ├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм
OUTPUT_ROOT = Path("/scratch/users/arunps/hindibabynet/audio_classified")

# Pick a random participant (or set manually)
available = sorted(
    d.name for d in OUTPUT_ROOT.iterdir()
    if d.is_dir() and not d.name.startswith("_")
)
assert available, f"No participant dirs found in {OUTPUT_ROOT}"

PARTICIPANT_ID = random.choice(available)
# PARTICIPANT_ID = "ABAN141223"  
# ├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм

participant_dir = OUTPUT_ROOT / PARTICIPANT_ID
pid = PARTICIPANT_ID

female_wav     = participant_dir / f"{pid}_main_female.wav"
male_wav       = participant_dir / f"{pid}_main_male.wav"
child_wav      = participant_dir / f"{pid}_child.wav"
background_wav = participant_dir / f"{pid}_background.wav"

wav_map = {
    "Main Female":  female_wav,
    "Main Male":    male_wav,
    "Child":        child_wav,
    "Background":   background_wav,
}

print(f"Randomly selected: {PARTICIPANT_ID}")
print(f"   Directory: {participant_dir}\n")
for label, wav in wav_map.items():
    status = "exists" if wav.exists() else "not found"
    print(f"   {label:15s} : {status}  ({wav.name})")
```

## Main Female Speaker

```python
def _play(wav_path):
    """Print info and display audio player for a WAV file."""
    if wav_path.exists():
        info = sf.info(wav_path)
        dur_min = info.duration / 60
        print(f"File     : {wav_path.name}")
        print(f"Duration : {info.duration:.1f} s  ({dur_min:.1f} min)")
        print(f"SR       : {info.samplerate} Hz  |  Channels: {info.channels}")
        print(f"Size     : {wav_path.stat().st_size / 1e6:.1f} MB")
        print()
        ipd.display(ipd.Audio(str(wav_path)))
    else:
        print(f"{wav_path.name} not found -- run Stage 03 first.")

_play(female_wav)
```

## Main Male Speaker

```python
_play(male_wav)
```

## Child

```python
_play(child_wav)
```

## Background

```python
_play(background_wav)
```

## Listen to a Specific Time Range

If the full files are too long, load and play a short clip from any of the four WAVs.

```python
# ├втАЭтВм├втАЭтВм Clip settings ├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм
WAV_TO_CLIP = female_wav       # or: male_wav, child_wav, background_wav
START_SEC   = 0.0              # start time in seconds
END_SEC     = 30.0             # end time in seconds
# ├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм├втАЭтВм

if WAV_TO_CLIP.exists():
    info = sf.info(WAV_TO_CLIP)
    start_frame = int(START_SEC * info.samplerate)
    n_frames    = int((END_SEC - START_SEC) * info.samplerate)
    audio, sr   = sf.read(WAV_TO_CLIP, start=start_frame, frames=n_frames, dtype="float32")
    print(f"Playing {WAV_TO_CLIP.name}  [{START_SEC:.1f}s - {END_SEC:.1f}s]  ({len(audio)/sr:.1f}s loaded)")
    ipd.display(ipd.Audio(audio, rate=sr))
else:
    print(f"{WAV_TO_CLIP} not found.")
```

## All Available Participants 

```python
print(f"Found {len(available)} participant(s) in {OUTPUT_ROOT}:\n")
for p in available:
    d = OUTPUT_ROOT / p
    fem  = (d / f"{p}_main_female.wav").exists()
    mal  = (d / f"{p}_main_male.wav").exists()
    chd  = (d / f"{p}_child.wav").exists()
    bg   = (d / f"{p}_background.wav").exists()
    marks = f"F={'Y' if fem else 'N'}  M={'Y' if mal else 'N'}  C={'Y' if chd else 'N'}  B={'Y' if bg else 'N'}"
    sel = " <-- selected" if p == PARTICIPANT_ID else ""
    print(f"  {p}  {marks}{sel}")
```
