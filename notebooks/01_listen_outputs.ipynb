# ğŸ§ Listen to Pipeline Outputs

Play back `main_female.wav` and `main_male.wav` produced by **Stage 03 â€” Speaker Classification**.

Set `PARTICIPANT_ID` below to switch between participants.

```python
from pathlib import Path
import IPython.display as ipd
import soundfile as sf

# â”€â”€ Configuration â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
OUTPUT_ROOT = Path("/scratch/users/arunps/hindibabynet/audio_classified")
PARTICIPANT_ID = "ABAN141223"  
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

participant_dir = OUTPUT_ROOT / PARTICIPANT_ID
female_wav = participant_dir / "main_female.wav"
male_wav   = participant_dir / "main_male.wav"

print(f"Participant : {PARTICIPANT_ID}")
print(f"Directory   : {participant_dir}")
print(f"Female WAV  : {'exists' if female_wav.exists() else ' not found'}")
print(f"Male WAV    : {'exists' if male_wav.exists() else ' not found'}")
```

## Main Female Speaker

```python
if female_wav.exists():
    info = sf.info(female_wav)
    duration_min = info.duration / 60
    print(f"Duration : {info.duration:.1f} s  ({duration_min:.1f} min)")
    print(f"SR       : {info.samplerate} Hz")
    print(f"Channels : {info.channels}")
    print(f"Size     : {female_wav.stat().st_size / 1e6:.1f} MB")
    print()
    ipd.display(ipd.Audio(str(female_wav)))
else:
    print(f"{female_wav} not found â€” run Stage 03 first.")
```

## Main Male Speaker

```python
if male_wav.exists():
    info = sf.info(male_wav)
    duration_min = info.duration / 60
    print(f"Duration : {info.duration:.1f} s  ({duration_min:.1f} min)")
    print(f"SR       : {info.samplerate} Hz")
    print(f"Channels : {info.channels}")
    print(f"Size     : {male_wav.stat().st_size / 1e6:.1f} MB")
    print()
    ipd.display(ipd.Audio(str(male_wav)))
else:
    print(f"{male_wav} not found â€” run Stage 03 first.")
```

## Listen to a Specific Time Range

If the full files are too long, load and play a short clip.

```python
# â”€â”€ Clip settings â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
WAV_TO_CLIP = female_wav       # or male_wav
START_SEC   = 0.0              # start time in seconds
END_SEC     = 30.0             # end time in seconds
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

if WAV_TO_CLIP.exists():
    info = sf.info(WAV_TO_CLIP)
    start_frame = int(START_SEC * info.samplerate)
    n_frames    = int((END_SEC - START_SEC) * info.samplerate)
    audio, sr   = sf.read(WAV_TO_CLIP, start=start_frame, frames=n_frames, dtype="float32")
    print(f"Playing {WAV_TO_CLIP.name}  [{START_SEC:.1f}s â€“ {END_SEC:.1f}s]  ({len(audio)/sr:.1f}s loaded)")
    ipd.display(ipd.Audio(audio, rate=sr))
else:
    print(f"{WAV_TO_CLIP} not found.")
```
