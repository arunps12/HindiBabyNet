```python
from pathlib import Path
import pandas as pd

ARTIFACTS_ROOT = Path("artifacts/runs")
```

```python
# list available runs
runs = sorted([p.name for p in ARTIFACTS_ROOT.iterdir() if p.is_dir()])
runs
```

```python
RUN_ID = runs[-1]  
RUN_DIR = ARTIFACTS_ROOT / RUN_ID
RUN_DIR
```

```python
rec_path = RUN_DIR / "data_ingestion" / "recordings.parquet"
assert rec_path.exists(), f"Missing: {rec_path}"

recordings = pd.read_parquet(rec_path)
recordings.head()
```

```python
recordings.info()
```

```python
recordings.describe(include="all")
```

```python
recordings.groupby("participant_id").agg(
    n_recordings=("recording_id", "count"),
    total_hours=("duration_sec", lambda x: x.sum() / 3600),
)
```

```python
recordings.groupby(
    ["participant_id", "session_date"]
).agg(
    n_files=("recording_id", "count"),
    hours=("duration_sec", lambda x: x.sum() / 3600),
).reset_index()
```

```python
recordings.assign(
    duration_hr=recordings["duration_sec"] / 3600
).sort_values("duration_hr", ascending=False).head(10)
```
